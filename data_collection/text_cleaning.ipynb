# delete numbers/characters

```python
from typing import Dict, List
from collections import OrderedDict
```

```python
with open(SLD+'/text_corpora/supported_substrings_frequency.json', 'r') as f:
    raw_data: Dict[str, int] = json.load(f)
len(raw_data)
```

```python
def all_numbers(tokens:List[str], tags:List):
    return all(tag in {slt.text.Tags.NUMBER, slt.text.Tags.PUNCTUATION, slt.text.Tags.SPACE} for tag in tags)

def all_characters(tokens:List[str], tags:List):
    return all(len(t)==1 for t in tokens)

def bad_length(tokens:List[str], tags:List):
    return len([t for t in tokens if t !=" "])<=1

line_filters = [
    all_numbers,
    all_characters,
    bad_length
]
```

```python
filtered = {
    # (line, count)
    line:count
    for line, count in tqdm(raw_data.items())
    if not all(word.isnumeric() or word.isupper() or len(word)==1  for word in re.sub(r"[,!\.٫?،]", " ", line).split())
}
len(raw_data), len(filtered), filtered
```

```python
filtered = sorted(filtered.items(), key=lambda item: (item[0].count(" "), item[1], item[0]), reverse=True)
```

```python
with open(SLD+'/text_corpora/supported_substrings_frequency.json', 'w') as f:
    json.dump(OrderedDict(filtered), f, indent=2, ensure_ascii=False)
len(filtered)
```
