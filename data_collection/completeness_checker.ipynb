```python
hfadbook1_path = os.path.join(drive_sign_recordings_dir, 'reference_clips', "pk-hfad-1")
len(os.listdir(hfadbook1_path))
```

```python
label_repacements = {
    # "":
    # "",

    # "minute(unit)":
    #     "minute(unit-of-time)",

    # "monitor(noun)":
    #     "monitor(prefect)",

    # "second(unit)":
    #     "second(unit-of-time)",

    # "uniform(noun)":
    #     "uniform(costume)",

    # "آئی":
    #     "آئی(came)",

    # "وہ-آئی":
    #     "وہ-آئی(came)",

    # "وہ-آیا":
    #     "وہ-آیا(came)",

    # "انارکلی":
    #     "انارکلی(جگہ)",

    # "شریف":
    #     "شریف(honorable)",

    # "عصر":
    #     "عصر(نماز)",

    # "ہم":
    #     "ہم(we)",

    # "ہم-آئیں":
    #     "ہم(we)-آئیں",

    # "ہم-جائیں-گے":
    #     "ہم(we)-جائیں-گے",

    # "پرسوں":
    #     "پرسوں(day-after-tomorrow)",

    # "چکا":
    #     "چکا(has)",

    # "کرتے":
    #     "کرتے(do)",

    # "گا":
    #     "گا(will)",

    # "چکے":
    #     "چکے(have)",

    # "اللہ-ہم-پر-رحم-amen":
    # "اللہ-ہم(we)-پر-رحم-amen",

    "کرتی":
        "کرتی(does)",

    "کرتا":
        "کرتا(does)",

}
print('\n'.join(label_repacements.keys()))
print()
print('\n'.join(label_repacements.values()))
```

```python
changed_counts = 0
for filename in (os.listdir(hfadbook1_path)):
    old_path = os.path.join(hfadbook1_path, filename)

    label, ext = filename.split('.')

    if label not in label_repacements:
        continue

    new_name = label_repacements[label] + '.' + ext.lower()
    new_path = os.path.join(hfadbook1_path, new_name)

    if not os.path.exists(old_path):
        raise FileNotFoundError(f'{old_path} does not exist. cannot rename.')
    if os.path.exists(new_path):
        if os.path.samefile(old_path, new_path):
            print(f'{filename} & {new_name} both already point to same file but have different string names, rename function might not work')
            input('continue?')
        else:
            raise FileExistsError(f'{new_path} already exists. cannot overwrite.')

    changed_counts +=1
    print(old_path, '\n',new_path, '\n\n')

    # os.rename(old_path, new_path)

f'{changed_counts}/{len(label_repacements)}'
```
